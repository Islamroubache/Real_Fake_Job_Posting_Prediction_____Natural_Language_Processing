```python
import re, string
import nltk
import numpy as np
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords


nltk.download('wordnet')
nltk.download('punkt_tab')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

texts_cleaned = []

# Regular expression for detecting emojis
emoji_pattern = re.compile(
    "[\U00010000-\U0010FFFF"  # Matches various emojis
    "\u2640-\u2642"            # Female and male signs
    "\u2600-\u26FF"            # Miscellaneous symbols
    "\u2700-\u27BF"            # Dingbats
    "\u2300-\u23FF"            # Miscellaneous technical symbols
    "\u2B50"                   # Star emoji
    "\u203C"                   # Double exclamation mark
    "\u2049"                   # Exclamation question mark
    "]+", flags=re.UNICODE)

for t in df.text:
    # Clean emojis from text
    t = re.sub(emoji_pattern, r"", t)

    # Remove \r and \n, lowercase text
    t = t.replace('\r', '').replace('\n', ' ').lower()

    # Remove links and mentions
    t = re.sub(r'http\S+', '', t)
    t = re.sub(r'@\S+', '', t)
    t = re.sub(r'#\S+', '', t)

    # # Remove non-UTF-8 characters
    t = re.sub(r'[^\x00-\x7f]',r'', t)

    # # Remove punctuation
    banned_list = string.punctuation
    table = str.maketrans('', '', banned_list)
    t = t.translate(table)

    # Remove stopwords
    t = [word for word in t.split() if word not in stop_words]
    t = ' '.join(t)

    # Remove words longer than 14 characters
    t = ' '.join(word for word in t.split() if len(word) < 14)

    # Remove contractions
    t = re.sub(r"can\'t", "can not", t)
    t = re.sub(r"n\'t", " not", t)
    t = re.sub(r"\'re", " are", t)
    t = re.sub(r"\'s", " is", t)
    t = re.sub(r"\'d", " would", t)
    t = re.sub(r"\'ll", " will", t)
    t = re.sub(r"\'t", " not", t)
    t = re.sub(r"\'ve", " have", t)
    t = re.sub(r"\'m", " am", t)

    # # Remove special characters like "&" and "$"
    t = ' '.join([word for word in t.split(' ') if ('$' not in word) and ('&' not in word)])

    # Remove multiple spaces
    t = re.sub("\s\s+", " ", t)

    # Lemmatization instead of stemming
    tokenized = nltk.word_tokenize(t)
    lemmatizer = WordNetLemmatizer()
    t = ' '.join([lemmatizer.lemmatize(word) for word in tokenized])  # Lemmatize each word

    # Append cleaned text to the list
    texts_cleaned.append(t)
```

```python
df['text_clean'] = texts_cleaned
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.2,  stratify=y_train, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

clf = LogisticRegression(max_iter = 1000)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_train)
accuracy_train_log_reg = accuracy_score(y_train, y_pred)

y_pred = clf.predict(X_valid)
accuracy_valid_log_reg = accuracy_score(y_valid, y_pred)

y_pred = clf.predict(X_test)
accuracy_test_log_reg = accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Train a Random Forest model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

y_pred = rf.predict(X_train)
accuracy_train_rf = accuracy_score(y_train, y_pred)

y_pred = rf.predict(X_valid)
accuracy_valid_rf = accuracy_score(y_valid, y_pred)

y_pred = rf.predict(X_test)
accuracy_test_rf = accuracy_score(y_test, y_pred)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Train an SVM model
svm = SVC(random_state=42)
svm.fit(X_train, y_train)

y_pred = svm.predict(X_train)
accuracy_train_svm = accuracy_score(y_train, y_pred)

y_pred = svm.predict(X_valid)
accuracy_valid_svm = accuracy_score(y_valid, y_pred)

y_pred = svm.predict(X_test)
accuracy_test_svm = accuracy_score(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Accuracy values (replace these with the actual accuracy values for your models)
models = ['Logistic Regression', 'SVM', 'Random Forest']
train_acc = [accuracy_train_log_reg, accuracy_train_svm, accuracy_train_rf]  # Replace with actual train accuracies
valid_acc = [accuracy_valid_log_reg, accuracy_valid_svm, accuracy_valid_rf]  # Replace with actual valid accuracies
test_acc = [accuracy_test_log_reg, accuracy_test_svm, accuracy_test_rf]  # Replace with actual test accuracies

# Set positions for the bars on the x-axis
x = np.arange(len(models))

# Width of the bars
width = 0.2

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Create bars for each dataset (Train, Validation, Test)
rects1 = ax.bar(x - width, train_acc, width, label='Train')
rects2 = ax.bar(x, valid_acc, width, label='Validation')
rects3 = ax.bar(x + width, test_acc, width, label='Test')

# Add labels and title
ax.set_xlabel('Models')
ax.set_ylabel('Accuracy')
ax.set_title('Model Accuracies on Train, Validation, and Test Datasets')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Display the value labels on top of each bar
def add_labels(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.4f}',  # Format to 4 decimal places
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Add labels to the bars
add_labels(rects1)
add_labels(rects2)
add_labels(rects3)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import re
import string
import nltk
import joblib
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Initialize NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Regular expression for cleaning emojis
emoji_pattern = re.compile(
    "[\U00010000-\U0010FFFF"
    "\u2640-\u2642"
    "\u2600-\u26FF"
    "\u2700-\u27BF"
    "\u2300-\u23FF"
    "\u2B50"
    "\u203C"
    "\u2049"
    "]+",
    flags=re.UNICODE
)

def clean_text(text):
    text = re.sub(emoji_pattern, "", text)  # Remove emojis
    text = text.replace('\r', '').replace('\n', ' ').lower()  # Clean line breaks and lowercase
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'@\S+', '', text)  # Remove mentions
    text = re.sub(r'#\S+', '', text)  # Remove hashtags
    text = re.sub(r'[^\x00-\x7f]', '', text)  # Remove non-ASCII characters
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    text = [word for word in text.split() if word not in stop_words]  # Remove stopwords
    text = ' '.join(text)
    text = ' '.join(word for word in text.split() if len(word) < 14)  # Remove long words
    text = re.sub(r"can\'t", "can not", text)  # Expand contractions
    text = re.sub(r"n\'t", " not", text)  # Expand contractions
    text = re.sub(r"\'re", " are", text)  # Expand contractions
    text = re.sub(r"\'s", " is", text)  # Expand contractions
    text = re.sub(r"\'d", " would", text)  # Expand contractions
    text = re.sub(r"\'ll", " will", text)  # Expand contractions
    text = re.sub(r"\'t", " not", text)  # Expand contractions
    text = re.sub(r"\'ve", " have", text)  # Expand contractions
    text = re.sub(r"\'m", " am", text)  # Expand contractions
    text = ' '.join(word for word in text.split() if ('$' not in word) and ('&' not in word))  # Remove unwanted symbols
    text = re.sub("\s\s+", " ", text)  # Remove extra spaces
    tokenized = nltk.word_tokenize(text)  # Tokenize text
    text = ' '.join([lemmatizer.lemmatize(word) for word in tokenized])  # Lemmatize words
    return text

# Load the trained pipeline model (includes vectorizer and classifier)
best_model = joblib.load('best_model.pkl')

# Test with a sample input
sample_text = "The job posting for software engineer is fraudulent!"

# Clean the sample text using the clean_text function
cleaned_text = clean_text(sample_text)

# Get the prediction directly from the model
prediction = best_model.predict([cleaned_text])  # Pass the cleaned text to the model

# Output the result
print(f'Cleaned Text: {cleaned_text}')
print(f'Prediction: {prediction[0]}')  # This will print the predicted class (fraudulent or not)

```
